```python
import pandas
```

```python
#importing excel file 
#data is taken from NBRM (National Bank of Republic of Macedonia) for the period of 30.03.20 to 30.03.23

usd_rates = pandas.read_excel("KursnaLista.xlsx")
```

```python
#visualization of data by day groups

usd_rates.Middle.plot()
```

```python
# we are going to make a simple test to check if we are going to get right data
# first we are going to take two dates (past 7 days, future 7 days)
# we are going to take days from the past to check how the dollar rate changed 
# based on that we can predict how the rate is going to change in the future (for e.g. next 7 days)

past = 7
future = 7
```

```python
money = usd_rates.Middle #get all rates
```

```python
money #print to test if we are getting the right data
```

```python
# now we are going to take ssome random day
# let it be 100 day
# now we are going ot take past 7 days from 100 day and 7 future days from 100 day

money[100-past:100]
```

```python
# 7 future days
money[100:100+future]
```

```python
# now we have some data set on which our model is going to learn
# it will now the past (in our e.g. 7 days and knows the future next 7 days)
# we have data and our model based on this data is going to learn how the rate will go

# now we want to take some point to start making test
# we can't get very first value because we will not know the past
# we also can't get the latest value as we don't know the future

# how much data we have
len(money)
```

```python
# taking data set
start = past
end = len(money) - future
print(f"We have {len(money)} values, and we will lok from {past} till {end} value")
```

```python
table = []

# for each of those days we are going to take past and future
for i in range(start,end) :
    s = money[(i-past):(i+future)]
    table.append(list(s))
    
```

```python
# preparing the data frame

past_columns = []
for i in range(past):
    past_columns.append(f"past_{i}")

    
future_columns = []
for i in range(future):
    future_columns.append(f"future_{i}")
    
```

```python
# creating table
data_frame = pandas.DataFrame(data = table, columns = (past_columns + future_columns)) 
```

```python
data_frame
```

```python
X = data_frame[past_columns] # getting input values on which we make prediction
X
```

```python
Y = data_frame[future_columns] # getting target values
Y
```

```python
# now we are going to devide this sets of data
# we will hide some litle amount of data in order to check if the model is correct
# we will show this hidden data after all tests

X = data_frame[past_columns][:10]
Y = data_frame[future_columns][:10]
```

```python
# hidden data

X_test = data_frame[past_columns][-10:] 
Y_test = data_frame[future_columns][-10:]
```

```python
pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression() #prepearing for learning
LR.fit(X,Y)
```

```python
prediction = LR.predict(X_test.head(1)) # checking how model is going to predict based on small amount of data
prediction
```

```python
Y_test.head(1) #checking what we except
```

```python
#visualizing data
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
```

```python
plt.plot(prediction[0], label = "Predicton")
plt.plot(Y_test.iloc[0], label = "Real")
plt.legend()
```

```python
# now we are going to see the results for every of our test values

for i in range(10):
    err = mean_absolute_error(prediction[0], Y_test.iloc[i])
    print(f"date {i}, error = {err}")
    prediction = LR.predict([X_test.iloc[i]])
    plt.plot(prediction[0], label = "Predicton")
    plt.plot(Y_test.iloc[i], label = "Real")
    plt.legend()
    plt.show()
          
```

```python
# as this model does not give us correct values we are going to use new one
```

```python
from sklearn.neighbors import KNeighborsRegressor #Метод ближайших соседей
KNN = KNeighborsRegressor()
KNN.fit(X,Y)
```

```python
for i in range(10):
    err = mean_absolute_error(prediction[0], Y_test.iloc[i])
    print(f"date {i}, error = {err}")
    prediction = KNN.predict([X_test.iloc[i]])
    plt.plot(prediction[0], label = "Predicton")
    plt.plot(Y_test.iloc[i], label = "Real")
    plt.legend()
    plt.show()
```

```python
# we can specify number of neighbours to get relevant data

KNN = KNeighborsRegressor(n_neighbors=1)
KNN.fit(X,Y)
```

```python
for i in range(10):
    err = mean_absolute_error(prediction[0], Y_test.iloc[i])
    print(f"date {i}, error = {err}")
    prediction = KNN.predict([X_test.iloc[i]])
    plt.plot(prediction[0], label = "Predicton")
    plt.plot(Y_test.iloc[i], label = "Real")
    plt.legend()
    plt.show()
```

```python
# now we will try new model

from sklearn.neural_network import MLPRegressor # using neural network
#MLPR = MLPRegressor(hidden_layer_sizes=(256,128,64,32), max_iter=10000, random_state = 42)
mlp = MLPRegressor(random_state=30, max_iter = 20000, hidden_layer_sizes=(256,128,64,32), activation = 'relu')
mlp.fit(X, Y)
mlp_prediction = mlp.predict(X_test)[0]

plt.plot(mlp_prediction, label="MLP Prediction")
plt.plot(Y_test.iloc[0], label="Real data")
plt.legend()

mae = mean_absolute_error(Y_test.iloc[0], mlp_prediction)
print(f"MLP MAE = {mae}")
```

```python
# Using Grid Search CV

#Reason: Model should give the minimal absolute error
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from numpy import linspace
```

```python
linspace(0.001, 0.01, 5)
```

```python
model = MLPRegressor(random_state=30)
param_grid = {
        "max_iter": [5000, 10000, 20000],
        "learning_rate_init": linspace(0.001, 0.01, 5),
        "activation": ["logistic", "relu"]
}

GS = GridSearchCV (model, param_grid, scoring="neg_mean_absolute_error", cv=3)

# doing cross validation - in our case grid search will train 90 models and will decide which fits the best
```

```python
GS.fit(X,Y)
```

```python
GS.best_params_
```

```python
best_model = GS.best_estimator_
```

```python
mlp = best_model
mlp.fit(X, Y)
mlp_prediction = mlp.predict(X_test)[0]

plt.plot(mlp_prediction, label="MLP Prediction")
plt.plot(Y_test.iloc[0], label="Real data")
plt.legend()

mae = mean_absolute_error(Y_test.iloc[0], mlp_prediction)
print(f"MLP MAE = {mae}")
```
